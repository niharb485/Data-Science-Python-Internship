## Objective

The goal of this project is to analyze the Titanic dataset using data visualization techniques and communicate important insights through charts and concise interpretations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.drop(columns='Cabin', inplace=True)
```

```python
df.isnull().sum()
```

```python
bins = [0, 12, 18, 35, 60, 100]
labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']

df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()
```

Insight: Female passengers had a much higher survival rate compared to male passengers.

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

Insight: First-class passengers had the highest survival chances.

```python
plt.figure(figsize=(8,4))
plt.hist(df['Age'], bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

Insight: Most passengers were young adults between ages 20–40.

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

Insight: Passenger class and fare showed strong relationships with survival.

```python
plt.figure(figsize=(8,4))
sns.barplot(x='FamilySize', y='Survived', data=df)
plt.title('Survival by Family Size')
plt.show()
```

Insight: Medium-sized families had better survival rates than very large families or solo travelers.

```python
sns.catplot(col="Sex", x="Pclass", y="Survived", kind="bar", data=df)
plt.show()
```

Insight: Female passengers consistently had higher survival rates across all passenger classes.

# Conclusion

This dashboard highlighted important survival patterns in the Titanic dataset. Gender, passenger class, age, and family size significantly influenced survival chances. Data visualization helped uncover these trends clearly and effectively.
